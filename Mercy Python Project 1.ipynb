```python
# 2.IQR Method
data=pd.read_csv("AmesHousing.csv")
Q1=data["SalePrice"].quantile(0.25)
Q3=data["SalePrice"].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q1+1.5*IQR
data=data[(data["SalePrice"]>=lower_bound) & (data["SalePrice"]<=upper_bound)]
print("outliers removed successfully!")
```

# Part 5:Simple Linear Regression

```python
# Fitting The Model(statsmodel)
# Fitting the model.
import statsmodels.api as sm
data.head()
```

```python
# Define the independent and dependent variables
X=sm.add_constant(data['Gr Liv Area'])      # Independent Variable
y=data['SalePrice']                         # Dependent Variable
# Fit the Simple Linear Regression Model.
model=sm.OLS(y,X).fit()
y_pred=model.predict(X)
# print the summary of the model
print(model.summary())
```

## Interpreting the Co-efficient,p-value and R-squared;
## *Coefficient(Gr Liv Area)---How much SalePrice increases for every 1 unit increase in Living Area.
## *p-value----if Gr Liv Area is a significant predictor of SalePrice.
## *R-squared----how much variation in SalePrice is explained by Gr Liv Area    

## coefficient = +62.3520                       p-value = (p<0.001--highly significant)          r-squared = 0.340(34% of variation in SalePrice is predicted by Gr Living Area
## for every one unit increase                   ;p<0.05(significant)                              i.e 34% of differences in House Price acn be explained by the Ground Living Area.                                                                
## in GroundLivingArea,                           >0.05(notsignificant)         
## increases by 62.3520 units             *The Relationship between SalePrice 
##                                         and Gr Liv Area is highly ignificant.

```python
# Plotting the Regression Line.
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.scatter(X['Gr Liv Area'],y,color='purple',label='Actual Data')
plt.plot(X['Gr Liv Area'],y_pred,color='blue',label='Regression line')
plt.xlabel('Gr Liv Area')
plt.ylabel('SalePrice')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

## The Relationship is linear,hence an increase in Gr Liv Area causes an inrease in SalePrice.

## Part 6:Multiple Linear Regression

```python
import statsmodels.api as sm   # Fitting The Multiple Regression Model.
x=data[['Gr Liv Area','Overall Qual','Garage Area','Total Bsmt SF','Year Built']]     # Independent Variables .
y=data['SalePrice']          # Dependent Variable

# Drop rows with NaN values in the independent variables
x_cleaned = x.dropna()
y_cleaned = y[x_cleaned.index] # Ensure y aligns with x after dropping rows

x_cleaned=sm.add_constant(x_cleaned)
model=sm.OLS(y_cleaned,x_cleaned).fit()
predictions=model.predict(x_cleaned)
print_model=model.summary()
print(print_model)
```
